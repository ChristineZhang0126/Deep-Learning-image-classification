```python
class Cutout(object):
    import numpy as np

class Cutout(object):
    def __call__(self, img, cutout_width):
        h = img.shape[0]
        w = img.shape[1]

        mask = np.ones(img.shape)

        y = np.random.randint(h)
        x = np.random.randint(w)

        x1 = np.clip(x - cutout_width // 2, 0, w)
        x2 = np.clip(x + cutout_width // 2, 0, w)

        y1 = np.clip(y - cutout_width // 2, 0, h)
        y2 = np.clip(y + cutout_width // 2, 0, h)

        mask[y1:y2, x1:x2] = 0.0

        img = img * mask

        return img

```

```python
# cutout = Cutout(50)
# plt.imshow(cutout(selected_image_2))
# plt.suptitle("Cutout example",fontsize=15)

# plt.show()

import numpy as np
import cv2
from google.colab.patches import cv2_imshow
import matplotlib.pyplot as plt


data = np.load("./DDD/data/task4_X_train.npy")

# Select the index of the image you want to work with
image_index = 4  # Change this to the index of the image you want to use

# Get the selected image
selected_image_2 = data[image_index] / 255.0  # Normalize the image to [0, 1]
cutout_width = 30
# Create the Cutout transformation object
cutout = Cutout()

# Apply the Cutout transformation to the selected image
cutout_image = cutout(selected_image_2.copy(), cutout_width)

# Display the original and cutout images using matplotlib
fig, (ax1, ax2) = plt.subplots(1, 2)

ax1.imshow(selected_image_2)
ax1.set_title("Original")

ax2.imshow(cutout_image)
ax2.set_title("Cutout")

plt.suptitle("Cutout example", fontsize=15)

plt.show()
```
